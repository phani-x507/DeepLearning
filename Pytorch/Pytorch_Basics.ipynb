# Matrix Multiplication
 Two main ways of performing Matrix Multiplication
 1. Element wise
 2. Matrix Multiplication (Dot Product)

```python
# Element Wise.
tensor = torch.tensor([1,2,3])
tensor * tensor
#1*1, 2*2, 3*3
```

```python
# Matrix Multiplication using torch.matmul()
torch.matmul(tensor,tensor)
# 1*1+2*2+3*3 = 14
```

```python
ones = np.ones((3,3))
ones
```

```python
tensor = torch.tensor(ones)
tensor
```

```python
tensor2 = torch.tensor([[1,2,3],[4,5,6]],dtype=torch.float64)
tensor2
```

```python

#Two dimensions should be equal to perform this
# if you do dotproduct then shape of column of 1st matrix should be equal to
# shape of rows on second matrix ex: 3x2 * 2*3
# if shape is same it works too
torch.matmul(tensor2,tensor)
```

```python
# Transpose
# we can do this if changing dimension helps to make dotproduct possible
print(f"Tensor1 Shape : {tensor.shape} and Tensor2 Shape : {tensor2.shape} ")
print(f"The inner dimensions doesnot match so, we can transpose")
print(f"Transposed Tensor2 = {tensor2.T} and shape is {tensor2.T.shape}")
print(f"Now Multiplying these two will get \n {torch.matmul(tensor,tensor2.T)}")
```

# Reshaping, Stacking, Squeezing, and Unsqueezing Tensors
  - Reshaping : Reshapes the tensor to different dimensions
  - View : Returns the view of a original tensor
  - Stacking : Combining Multiple Tensors
  - Squeeze : Removes all 1 Dimension From a tensor
  - Unsqueeze : Add a 1 dimension to a target tensor
  - Permute : Return a view of the input with dimensions permuted(Swapped) in a certain way

```python
x = torch.arange(1,10)
x
```

```python
# Now the possible reshapes are the factors of total items
torch.reshape(x,(3,3))
```

```python
torch.reshape(x,(1,9))
```

```python
torch.reshape(x,(9,1))
```

```python
# Adds Another Dimension but directs to same memory of original
# Change in value of a view changes in original array
z = x.view(3,3)
z[1,0] = 55
z," ",x
```

```python
#Stacking means Adding elements to existing matix horizontally or vertically
# stack(),hstack(),vstack()
# Creates another dimension and does manipulation
# dim = 0 (row) -> Vstack() , dim = 1(column) -> hstack(), dim = 0 by default
y = torch.arange(0,90,10)
y = y.reshape(3,3)
y_stacked = torch.stack([z,y],dim=1)
y_stacked
```

```python
x = x.view(1,9)
x,'Original Dimension',x.shape
```

```python
x.squeeze(),'After Squeezing',x.squeeze().shape # Removes all single Dimensions
```

```python
x.unsqueeze(dim=1),'After Unsqueezing', x.unsqueeze(dim=1).shape #Adds 1 Dimension
```
