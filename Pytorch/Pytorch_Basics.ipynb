# Tensor Datatypes

# Float 32 tensor
**Note**: Tensor datatypes is one the 3 big errors you run into pythorch and deeplearning
1. Tensors not right datatype
2. Tensors not right ShapeType
3. Tensors not on the right device

```python
float_32_tensor = torch.tensor([3.0,6.0,9.0],
                             device = None, #CPU, Cuda or None(Defaults CPU)
                             dtype = None #torch.float32,16,64 or any.
                              )
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
# Convert Tensor
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor.dtype
```

```python
# Details of Tensor.
print("Data Type : ",float_16_tensor.dtype)
print("Shape of a Tensor :", float_16_tensor.shape )
print("Device tensor is on:", float_32_tensor.device)
```
