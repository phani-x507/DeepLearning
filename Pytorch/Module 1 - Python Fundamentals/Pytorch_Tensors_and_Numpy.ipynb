## Pytorch tensors And Numpy
Convert Numpy to pytorch tensor ->  `torch.from_numpy(ndarray)`
PyTorch tensor -> Numpy -> `torch.Tensor.numpy()`

```python
# Numpy array to tensor
import torch
import numpy as np

array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
array.dtype #Numpy creates float64 by default
```

```python
tensor.dtype # After conversion tensor adapts the dtype of numpy
```

```python
torch.arange(1.0,7.0).dtype #Tensor creates float32 by default
```

```python
#if we want to change it to float 32
tensor = torch.from_numpy(array).type(torch.float32)
tensor.dtype
```

```python
# Tensor to Numpy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
numpy_tensor.dtype
```

# Reproducibility
* Reproducibility in PyTorch means getting the same results (or very close) every time you run your code—especially important for debugging, research, and comparing experiments.
* It is used to produce near to similar random numbers everytime to get same randomness in sample data every time running the code.

```python
## Reproducibility (Trying to pick random from random numbers )
tensorA = torch.rand(3,3)
tensorB = torch.rand(3,3)
print(tensorA)
print(tensorB)
print(tensorA == tensorB)
```

```python
## We can use manual seed for reproducibility
RANDOM_SEED = 43
torch.manual_seed(RANDOM_SEED)
tensorA = torch.rand(3,3)
tensorB = torch.rand(3,3)
print(tensorA == tensorB) # So tensorA and tensorB is not created same results
```

```python
# Lets to like this
torch.manual_seed(RANDOM_SEED)
tensorA = torch.rand(3,3)
torch.manual_seed(RANDOM_SEED)
tensorB = torch.rand(3,3)
print(tensorA == tensorB) # it created same results
```

```python
# Device Agnoustic Code
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
# Count number of cuda Devices
torch.cuda.device_count()
```

```python
# Lets create a tensor
tensor = torch.rand(3)
tensor.device #So by default tensor is created by cpu
```

```python
#Lets move it to GPU
tensorGPU = tensor.to(device)
tensorGPU # Here we successfully converted to GPU
```

```python
# Now Convert into NumPy array
# tensorGPU.numpy() -> it shows error,
#can't convert cuda:0 device type tensor to numpy.
#Use Tensor.cpu() to copy the tensor to host memory first.

# So we need to shift that to cpu then convert it to numpy

tensorCPU = tensorGPU.cpu().numpy()
tensorCPU,tensorCPU.device #Conversion Successful
```
